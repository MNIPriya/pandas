```python
# To make this cell executable independently and ensure 'Age' column exists for filling
import pandas as pd
import numpy as np

# Re-defining the original data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000]
}
df = pd.DataFrame(data)

# filling the 'Age' column
# The original code was df["Age"].fillna(df["Age"].mean())
# For a persistent change in df, inplace=True is commonly used
#mathametically
df["Age"].fillna(df["Age"].mean())
df["Age"]
df["Salary"].fillna(df["Salary"].median())
df["Salary"]
#filling random values -> forward fill , -> bakward fill
df["Age"].fillna(method = "ffill")
df["Age"].fillna(method = "bfill")


```

```python
df["Name"].replace("Charlie","Roise")
```

```python
#deling with duplicate values
df_dup = df[df.duplicated(keep = "first")]
df_dup = df[df.duplicated(keep = "last")]
df_dup
```

```python
df = df.drop_duplicates()
df
```

```python
df["Salary"] = df["Salary"].apply(lambda x: x/10 if x > 6500000 else x )
df["Salary"]
#df[["first_name","last_name"]] = df["name"].str.split(_)
#apply , lambda
def muliply(x):
  return x*2

df["Age"] = df["Age"].apply(muliply)
df["Age"]
```
